```python
import pymysql

# Connect once
conn = pymysql.connect(
    user='root',
    password='root',
    host='localhost',
    database='rest_review_db',
    autocommit=True 
)

foods = [
    "Idly", "Dosa", "Vada", "Roti", "Meals", "Veg Biryani",
    "Egg Biryani", "Chicken Biryani", "Mutton Biryani",
    "Ice Cream", "Noodles", "Manchurian", "Orange Juice",
    "Apple Juice", "Pineapple Juice", "Banana Juice"
]

# ✅ Safe query (no duplicate error)
qur = """
INSERT INTO reviews_table (food, good_review, bad_review, customer)
VALUES (%s, %s, %s, %s)
ON DUPLICATE KEY UPDATE 
    good_review = VALUES(good_review),
    bad_review = VALUES(bad_review),
    customer = VALUES(customer)
"""

mycur = conn.cursor()

# Loop
for i in foods:
    mycur.execute(qur, (i, 0, 0, 0))

conn.commit()

# Close
mycur.close()
conn.close()
```
